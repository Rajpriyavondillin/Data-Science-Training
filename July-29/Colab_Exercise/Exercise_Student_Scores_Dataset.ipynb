```python
# 1. Create a CSV and Read It

import pandas as pd

data={
    'Name': ['Amit', 'Priya', 'Kiran', 'Anita', 'Ravi'],
    'Math': [78, 92, 65, 88, 55],
    'Science': [85, 89, 70, 95, 60],
    'English': [74, 96, 60, 90, 58]
}

df = pd.DataFrame(data)

# Create a CSV in Colab's virtual environment (temporary storage):
df.to_csv('students.csv', index=False)
```

```python
# Read 'students.csv' file
df = pd.read_csv('/content/students.csv')

# Read it into a pandas DataFrame and print the first 3 rows.
print(df.head(3))
```

```python
# 2. Calculate Average Scores

# Add a new column Average that calculates the mean score for Math, Science, and English for each student.
df['Average'] = df[['Math','Science','English']].mean(axis=1)

# Print the student with the highest average.
print("\nStudent with highest average:")
top_student = df.loc[df['Average'].idxmax()]
print(top_student)
```

```python
# 3. Filter Students

# Display all students who have scored more than 80 in Science.
print("\nStudent scored > 80 in Science:\n ", df[df['Science'] > 80])

# Display all students with an average score below 70.
print("\nStudent with average score < 70: \n", df[df['Average'] < 70])
```

```python
# 4. Sorting

# Sort the students by English score (descending) and print the top 3.
print("\nTop 3 students by English score: \n", df.sort_values(by='English', ascending=False).head(3))
```

```python
# 5. Chart Exercise

# Create a bar chart showing studentsâ€™ names on the X-axis and their average scores on the Y-axis.
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.bar(df['Name'], df['Average'], color='skyblue')
plt.xlabel('Students')
plt.ylabel('Average Score')
plt.title('Students Average Scores')
plt.grid(True)
plt.show()


```

```python
# Create a line chart for all studentsâ€™ scores in Math, Science, and English (use df.plot() ).
df.set_index('Name', inplace=True)
df.plot(kind='line',marker='o',figsize=(8,5))

plt.title('Student Scores in Math, Science, and English')
plt.xlabel('Student Name')
plt.ylabel('Score')
plt.grid(True)
plt.legend(title='Subjects')
plt.tight_layout()
plt.show()
```

```python
# 6. Bonus Task

# Find the subject with the highest total score across all students.
subject_totals = df[['Math', 'Science', 'English']].sum()
print("\nSubject with highest total score:", subject_totals.idxmax())

# Save the updated DataFrame (with Average column) as a new CSV named students_summary.csv .
df.reset_index(inplace=True)
df.to_csv('students_summary.csv', index=False)
print('\nSaved as students_summary.csv')

print(df.columns)

# Download 'students_summary.csv' file in our local computer
files.download('students_summary.csv')
```
