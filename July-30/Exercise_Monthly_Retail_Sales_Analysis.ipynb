# Exercise: Monthly Retail Sales Analysis


```python
# 1. Create a CSV of Sales Data

import pandas as pd

data={
    'Month': ['Jan', 'Jan', 'Feb', 'Feb', 'Mar', 'Mar'],
    'Store': ['Store A', 'Store B','Store A', 'Store B','Store A', 'Store B'],
    'Sales': [12000,15000,18000,17000,16000,19000]
}

df = pd.DataFrame(data)

# Create a CSV in Colab's virtual environment (temporary storage):
df.to_csv('sales_data.csv', index=False)
```

```python
# Read 'students.csv' file
df = pd.read_csv('/content/sales_data.csv')

# Read it into a pandas DataFrame and print the first 3 rows.
print(df)
```

```python
# 2. Total Sales Per Store
# Group the data by Store and calculate total sales for each store.
total_sales_per_store = df.groupby('Store')['Sales'].sum()
print("\nTotal Sales Per Store:\n",total_sales_per_store)
```

```python
# Identify which store had higher total sales.
max_sales_store = total_sales_per_store.idxmax()
print(f"The store with the highest total sales is: {max_sales_store}")
```

```python
# 3. Monthly Sales Trend
# Group the data by Month and sum up the total sales (Store A + Store B).
monthly_sales = df.groupby('Month')['Sales'].sum()

# Print the result.
print("\nMonthly Sales Trend:\n",monthly_sales)
```

```python
# 4. Chart: Monthly Sales by Store
# Create a line chart showing monthly sales for both Store A and Store B.
import matplotlib.pyplot as plt
pivot_df = df.pivot(index='Month', columns='Store', values='Sales')
# Plot the line chart
ax = pivot_df.plot(kind='line', marker='o', grid=True)

# Customize the chart
ax.set_title('Monthly Sales by Store')
ax.set_xlabel('Month')
ax.set_ylabel('Sales')
ax.legend(title="Store")

# Display the chart
plt.tight_layout()
plt.show()
```

```python
# 5. Bonus
# Create a bar chart showing total sales per store.
# Add data labels on top of each bar.
ax = total_sales_per_store.plot(kind='bar', color=['skyblue', 'orange'])
plt.title("Total Sales Per Store")
plt.ylabel("Total Sales")
plt.xticks(rotation=0)

# Add data labels
for i, val in enumerate(total_sales_per_store):
    ax.text(i, val + 500, f'{val}', ha='center', va='bottom')

# Save this chart as an image ( sales_chart.png ).
plt.tight_layout()
plt.savefig("sales_chart.png")
plt.show()
```
